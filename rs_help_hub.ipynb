### Algoritmo para RS:

```python
def recomendar_perfiles(habilidad_a_aprender=None, dias_preferidos=None, rango_tiempo_preferido=None):
    filtros = True
    if habilidad_a_aprender:
        filtros &= perfil_df['category'].str.contains(habilidad_a_aprender, case=False, na=False)
    if dias_preferidos:
        filtros &= perfil_df['selectedDays'].str.contains(dias_preferidos, case=False, na=False)
    if rango_tiempo_preferido:
        filtros &= perfil_df['preferredTimeRange'].str.contains(rango_tiempo_preferido, case=False, na=False)

    perfiles_coincidentes = perfil_df[filtros]
    
    resúmenes_usuarios = []
    for idx in perfiles_coincidentes.index:
        info_usuario = usuarios_df.loc[usuarios_df.index[idx]].to_dict()
        resumen_usuario = {
            "nombre": info_usuario['nameUser'],
            "apellido": info_usuario['surnameUser'],
            "ubicación": perfiles_coincidentes.at[idx, 'location'],
            "descripción": perfiles_coincidentes.at[idx, 'description'],
            "foto_perfil": perfiles_coincidentes.at[idx, 'profilePicture'],
            "habilidades_interesadas": perfiles_coincidentes.at[idx, 'interestedHabilities']
        }
        resúmenes_usuarios.append(resumen_usuario)

    return perfiles_coincidentes, resúmenes_usuarios
```

```python
# Estructura para calificaciones
calificaciones_df = pd.DataFrame(columns=['user_id', 'skill', 'rating', 'type'])
```

```python
def calificar_habilidad(user_id, skill, rating, type='ofrecida'):
    global calificaciones_df
    
    if rating < 1 or rating > 5:
        print("La calificación debe estar entre 1 y 5.")
        return

    nueva_calificacion = {'user_id': user_id, 'skill': skill, 'rating': rating, 'type': type}
    calificaciones_df = calificaciones_df.append(nueva_calificacion, ignore_index=True)
    print(f"Calificación de {rating} para la habilidad '{skill}' registrada exitosamente.")
```

```python
def calcular_promedio_calificacion(skill, type='ofrecida'):
    calificaciones_filtradas = calificaciones_df[(calificaciones_df['skill'] == skill) & (calificaciones_df['type'] == type)]
    if not calificaciones_filtradas.empty:
        promedio = calificaciones_filtradas['rating'].mean()
        print(f"El promedio de calificación para la habilidad '{skill}' es: {promedio:.2f}")
        return promedio
    else:
        print(f"No hay calificaciones para la habilidad '{skill}'.")
        return None
```

### Algoritmo para Notificaciones Automáticas:

Para implementar un sistema de notificaciones automáticas cuando se registren nuevos perfiles que coincidan con las habilidades de un usuario, se sigue estos pasos:

- Crear una lista de usuarios y sus habilidades de interés.
- Al registrar un nuevo perfil, comprobar si ofrece habilidades que están en la lista de intereses de los usuarios.
- Enviar notificaciones a los usuarios interesados.

```python
def enviar_notificacion(email, mensaje):
    print(f'Notificación enviada a {email}: {mensaje}')

def registrar_perfil(nuevo_perfil):
    global perfil_df

    nuevo_perfil_df = pd.DataFrame([nuevo_perfil])
    perfil_df = pd.concat([perfil_df, nuevo_perfil_df], ignore_index=True)

    # Comprobar si el nuevo perfil ofrece habilidades que interesan a los usuarios
    for _, usuario in usuarios_df.iterrows():
        for habilidad in nuevo_perfil['offeredHabilities']:
            if habilidad in usuario['interestedHabilities']:
                mensaje = f"Se ha registrado un nuevo perfil que ofrece habilidades de: {habilidad}"
                enviar_notificacion(usuario['email'], mensaje)

    # Comprobar si el usuario del nuevo perfil tiene intereses que coincidan con habilidades ofrecidas por otros
    for _, perfil_existente in perfil_df.iterrows():
        for habilidad in perfil_existente['offeredHabilities']:
            if habilidad in nuevo_perfil['interestedHabilities']:
                mensaje = f"Un perfil existente ofrece una habilidad que te interesa: {habilidad}"
                enviar_notificacion(nuevo_perfil['email'], mensaje)

    print("Perfil registrado y las notificaciones fueron enviadas correctamente.")
```
